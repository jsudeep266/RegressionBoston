```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=load_boston()
```

```python
dataset= pd.DataFrame(df.data)
```

```python
x=dataset.iloc[:,:-2]
y=dataset.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lg=LinearRegression()
mse=cross_val_score(lg,x,y,scoring="neg_mean_squared_error",cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={"alpha":[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,32,40,45,50,55,100]}
rg=GridSearchCV(ridge,parameters,scoring="neg_mean_squared_error",cv=5)
rg.fit(x,y)
```

```python
print(rg.best_params_)
print(rg.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
parameters={"alpha":[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ls=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
ls.fit(x,y)
print(ls.best_params_)
print(ls.best_score_)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
pre_lasso=ls.predict(x_test)
pre_ridge=rg.predict(x_test)
```
